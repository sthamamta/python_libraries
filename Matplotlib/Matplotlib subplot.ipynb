```python
plt.subplot(3,2,1)  #3 rows 2 columns 1 st fig
plt.pie([1])

plt.subplot(3,2,3) #3 rows 2 columns 3rd fig
plt.pie([1,2])

plt.subplot(3,2,4) #3 rows 2 columns 4th fig
plt.pie([1,2,3])

plt.show()
```
