```python
import matplotlib.pyplot as plt
import numpy as np
import random
```

```python
ml_std_age = np.random.randint(18,50,(100))
py_std_age = np.random.randint(15,40,(100))
```

```python
print(ml_std_age)
```

```python
print(py_std_age)
```

```python
plt.hist(ml_std_age)
plt.title('Machine Learning Students age histogram')
plt.xlabel('Students age category')
plt.ylabel('no of stds')
plt.show()
```

```python
#using bins and rwidth
bins=[15,20,25,30,35,40,45,50]
plt.hist(ml_std_age,bins,rwidth=0.9)
plt.title('Machine Learning Students age histogram')
plt.xlabel('Students age category')
plt.ylabel('no of stds')
plt.show()
```

```python
#using histtype
bins=[15,20,25,30,35,40,45,50]
plt.hist(ml_std_age,bins,rwidth=0.9,histtype="step")
plt.title('Machine Learning Students age histogram')
plt.xlabel('Students age category')
plt.ylabel('no of stds')
plt.show()
```

```python
#using orientation
bins=[15,20,25,30,35,40,45,50]
plt.hist(ml_std_age,bins,rwidth=0.9,histtype="bar",align='left',orientation='horizontal')
plt.title('Machine Learning Students age histogram')
plt.xlabel('Students age category')
plt.ylabel('no of stds')
plt.show()
```

```python
#using color
bins=[15,20,25,30,35,40,45,50]
plt.hist(ml_std_age,bins,rwidth=0.9,histtype="bar",align='left',orientation='horizontal',color='m',label="ml student")
plt.title('Machine Learning Students age histogram')
plt.xlabel('Students age category')
plt.ylabel('no of stds')
plt.legend()
plt.show()
```

```python
#using multiple data plot
bins=[15,20,25,30,35,40,45,50]
plt.hist(ml_std_age,bins,rwidth=0.9,histtype="bar",orientation='vertical',color='m',label="ml student")
plt.hist(py_std_age,bins,rwidth=0.9,histtype="bar",orientation='horizontal',color='r',label="py student")
plt.title('Machine Learning Students age histogram')
plt.xlabel('Students age category')
plt.ylabel('no of stds')
plt.legend()
plt.show()
```

```python
bins=[15,20,25,30,35,40,45,50]
plt.hist([ml_std_age,py_std_age],bins,rwidth=0.9,histtype="bar",orientation='vertical',
         color=['m','b'],label=["ml student","Py student"])
#plt.hist(py_std_age,bins,rwidth=0.9,histtype="bar",orientation='horizontal',color='r',label="py student")
plt.title('Machine Learning Students age histogram')
plt.xlabel('Students age category')
plt.ylabel('no of stds')
plt.legend()
plt.show()
```

```python
#using figsize to increase the size of figure
bins=[15,20,25,30,35,40,45,50]
plt.figure(figsize=(16,9))
plt.hist([ml_std_age,py_std_age],bins,rwidth=0.9,histtype="bar",orientation='vertical',
         color=['m','b'],label=["ml student","Py student"])
#plt.hist(py_std_age,bins,rwidth=0.9,histtype="bar",orientation='horizontal',color='r',label="py student")
plt.title('Machine Learning Students age histogram')
plt.xlabel('Students age category')
plt.ylabel('no of stds')
plt.legend()
plt.show()
```

```python
#using style
from matplotlib import style
style.use("ggplot")
bins=[15,20,25,30,35,40,45,50]
plt.figure(figsize=(16,9))
plt.hist([ml_std_age,py_std_age],bins,rwidth=0.9,histtype="bar",orientation='vertical',
         color=['m','b'],label=["ml student","Py student"])
plt.title('Machine Learning Students age histogram')
plt.xlabel('Students age category')
plt.ylabel('no of stds')
plt.legend()
plt.show()
```
