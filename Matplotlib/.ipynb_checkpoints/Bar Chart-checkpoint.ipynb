```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
```

```python
classes = ["Python", "R", "AI", "ML", "DS"]
class1_students = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_students = [40, 5, 20, 20, 10]
class3_students = [35, 5, 30, 15, 15]
```

```python
#using bar
plt.bar(classes,class1_students)
plt.show()
```

```python
#horizontal bar
plt.barh(classes,class1_students)
```

```python
#using width
plt.bar(classes,class1_students,width=0.2)
```

```python
plt.bar(classes,class1_students,width=0.2,align='edge',color='m')
```

```python
#changing edge color and width
plt.bar(classes,class1_students,width=0.2,align='edge',color='m',edgecolor='b',linewidth=3)
```

```python
#changing the opacity with alpha parameter which takes value bet 0 and 1
plt.bar(classes,class1_students,width=0.2,align='edge',color='m',edgecolor='b',linewidth=3,alpha=0.2)
```

```python
#using linestyle
plt.bar(classes,class1_students,width=0.2,align='edge',color='m',edgecolor='b',linewidth=3,alpha=0.9,linestyle='--')
```

```python
#using label and legend
plt.bar(classes,class1_students,width=0.2,align='edge',color='m',edgecolor='b',linewidth=3,alpha=0.9,
        label='Class 1 student')
```

```python
# we can hide the graph by setting visible parameter to false
plt.bar(classes,class1_students,width=0.2,align='edge',color='m',edgecolor='b',linewidth=3,alpha=0.9,
        label='Class 1 student',visible=False)
```

# Part 2

```python
classes = ["Python", "R", "AI", "ML", "DS"]
class1_stds = [30, 10, 20, 25, 10] # out of 100 student in each class
class2_stds = [40, 5, 20, 20, 10]
class3_stds = [35, 5, 30, 15, 15]
```

```python
#using style for ggplot
style.use('ggplot')
plt.figure(figsize=(16,9))  #changing the figure size
plt.bar(classes,class1_stds,width=0.5,align='edge',color='k',edgecolor='m',linewidth=3,alpha=0.9,linestyle='--')
```

```python
style.use('ggplot')
plt.figure(figsize=(16,9))  #changing the figure size
plt.barh(classes,class1_stds,align='edge',color='k',edgecolor='m',linewidth=3,alpha=0.9,linestyle='--')
```

```python
style.use('ggplot')
plt.figure(figsize=(16,9))  #changing the figure size
plt.title('bar chart of students',fontsize=15)
plt.xlabel('classes',fontsize=12)
plt.ylabel('no of students',fontsize=12)
plt.bar(classes,class1_stds,width=0.5,align='edge',color='k',edgecolor='m',linewidth=5,alpha=0.9,linestyle='--')
```

```python
#showing multiple data on same bar
style.use('ggplot')
plt.figure(figsize=(14,7))  #changing the figure size
plt.title('bar chart of students',fontsize=15)
plt.xlabel('classes',fontsize=12)
plt.ylabel('no of students',fontsize=12)
plt.bar(classes,class1_stds,width=0.5,color='k')

plt.bar(classes,class2_stds,width=0.5,color='b')

```

```python
#showing multiple data on same bar
style.use('ggplot')
plt.figure(figsize=(14,7))  #changing the figure size

classes_index=np.arange(len(classes))
width = 0.2

plt.bar(classes_index,class1_stds,width,color='k',label='Class 1 students')
plt.bar(classes_index+width,class2_stds,width,color='b',label='class 2 students')

plt.xticks(classes_index+width,classes,rotation=30)

plt.legend()
plt.title('bar chart of students',fontsize=15)
plt.xlabel('classes',fontsize=12)
plt.ylabel('no of students',fontsize=12)
plt.show()

```

```python
style.use('ggplot')
plt.figure(figsize=(14,7))  #changing the figure size

classes_index=np.arange(len(classes))
width = 0.2

plt.bar(classes_index,class1_stds,width,color='k',label='Class 1 students')
plt.bar(classes_index+width,class2_stds,width,color='b',label='class 2 students')
plt.bar(classes_index+(2*width),class3_stds,width,color='m',label='class 3 students')

plt.xticks(classes_index+width,classes,rotation=30)

plt.legend()
plt.title('bar chart of students',fontsize=15)
plt.xlabel('classes',fontsize=12)
plt.ylabel('no of students',fontsize=12)
plt.show()

```

```python
style.use('ggplot')
plt.figure(figsize=(14,7))  #changing the figure size

classes_index=np.arange(len(classes))
width = 0.5

plt.bar(classes_index,class1_stds,width,color='k',label='Class 1 students')  #occupies 50% width
plt.bar(classes_index+width,class2_stds,width,color='b',label='class 2 students') #occupies remaining 50% width
plt.bar(classes_index+(2*width),class3_stds,width,color='m',label='class 3 students')  #overlaps with the next bar

plt.xticks(classes_index+width,classes,rotation=30)

plt.legend()
plt.title('bar chart of students',fontsize=15)
plt.xlabel('classes',fontsize=12)
plt.ylabel('no of students',fontsize=12)
plt.show()
```

```python
#horizontal bar
style.use('ggplot')
plt.figure(figsize=(14,7))  #changing the figure size

classes_index=np.arange(len(classes))
width = 0.2

plt.barh(classes_index,class1_stds,width,color='k',label='Class 1 students')
plt.barh(classes_index+width,class2_stds,width,color='b',label='class 2 students')
plt.barh(classes_index+(2*width),class3_stds,width,color='m',label='class 3 students')

# plt.xticks(classes_index+width,classes,rotation=30)
plt.yticks(classes_index+width,classes,rotation=30)

plt.legend()
plt.title('bar chart of students',fontsize=15)
plt.ylabel('classes',fontsize=12)
plt.ylabel('no of students',fontsize=12)
plt.show()
```
