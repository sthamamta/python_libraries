```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import style
import pandas as pd
```

```python
df = pd.read_csv('data/Google-Playstore.csv')
df.head()
```

```python
df.shape
```

```python
#taking only 1000 rows with nrows parameter

df1 = pd.read_csv('data/Google-Playstore.csv',nrows=1000)
df1
```

```python
df1["Reviews"]
```

```python
x = df1["Rating"]
y = df1["Reviews"]
```

```python
plt.scatter(x,y) # Draw scatter plot with parameter x and y
plt.title("Google Play Store Apps Scatter plot",fontsize=18)
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize = (16,9)) # figure size ratio 16:9
# plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha=0.5, linewidths=10,
#            edgecolors="g" )#verts="<"
plt.scatter(x,y, c = "r")#vcolor parameter
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize = (16,9)) # figure size ratio 16:9
plt.scatter(x,y, c = "r",marker='*', s=100) #changing the marker and marker size
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize = (16,9)) # figure size ratio 16:9
plt.scatter(x,y, c = "r",marker='*', s=100,alpha=0.3) #changing the transparency with alpha parameter
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize = (16,9)) # figure size ratio 16:9
plt.scatter(x,y, c = "r",marker='*', s=100,linewidths=10,edgecolors="g") #changing the linewidth and marker color
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize = (16,9)) # figure size ratio 16:9
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha=0.8, linewidths=10,
           edgecolors="g" )#verts="<"
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
plt.figure(figsize = (16,9)) # figure size ratio 16:9
plt.scatter(x,y, c = "r", s = 100, alpha=0.5, linewidths=2,
           edgecolors="g",verts="<" )#verts="<"
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews")
plt.show()
```

```python
#plotting two scatter plots
plt.figure(figsize = (16,9))
plt.scatter(x,y, c = "r", marker = "*", s = 100, alpha=0.5,
           edgecolors="g" )#verts="<"
 
plt.scatter(x,df1["Installs"], c = "y", marker = "o", s = 100, alpha=0.5, 
           edgecolors="c" )
plt.title("Google Play Store Apps Scatter plot")
plt.xlabel("Rating")
plt.ylabel("Reviews and Installs")
plt.show()
```
