```python
# import libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# missing value imputation using mean,median,mode
```

```python
df=pd.read_csv('data/train.csv')
df.head()
```

```python
df.shape
```

```python
pd.set_option('display.max_columns',None)
pd.set_option('display.max_rows',None)
```

```python
df.head()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
#calculating missing value percentage
missing_value_per = df.isnull().sum()/df.shape[0] *100
missing_value_per
```

```python
#displaying the feature with mising data percentage more than 20
drop_columns=missing_value_per[missing_value_per>20].keys()
drop_columns
```

```python
#eliminating the feature with mising data percentage more than 17
df2_drop_clm=df.drop(columns=drop_columns)
```

```python
df2_drop_clm.shape
```

```python
#extracting the rows having only datatype as int64 and float64
df3_num=df2_drop_clm.select_dtypes(include=['int64','float64'])
```

```python
#extracting dataframe containing only numeric variables
df3_num.head()
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(df3_num.isnull())
plt.show()
#white patches in heatmap represents the null value present in dataset
```

```python
#displaying the rows with null values
df3_num[df3_num.isnull().any(axis=1)]
```

```python
df3_num.isnull().sum()
```

```python
#creating the list of variable having null value in any rows in dataframe
missing_num_var =[var for var in df3_num.columns if df3_num[var].isnull().sum()>0]
missing_num_var
```

```python
plt.figure(figsize=(25,25))
for i,var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var],bins=20,kde_kws={'linewidth':4,'color':'m'})
    sns.distplot(df[var],bins=20)
```

```python
#when data is normally distributed then mean can be used to impute the missing value
```

```python
#when data is skewed ie not normally distributed then we use modian to impute the missing value
```

```python
#filling null values with mean value
df4_num_mean=df3_num.fillna(df3_num.mean())
df4_num_mean
```

```python
df4_num_mean.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df4_num_mean[var], bins=20, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    plt.legend()
```

# using median value to impute


```python
df5_num_median = df3_num.fillna(df3_num.median())
df5_num_median.isnull().sum().sum()
```

```python
plt.figure(figsize=(10,10))
sns.set()
for i, var in enumerate(missing_num_var):
    plt.subplot(2,2,i+1)
    sns.distplot(df3_num[var], bins=20,hist=False, kde_kws={'linewidth':8, 'color':'red'}, label="Original",)
    sns.distplot(df4_num_mean[var], bins=20,hist=False, kde_kws={'linewidth':5, 'color':'green'},label="Mean",)
    sns.distplot(df5_num_median[var], bins=20,hist=False, kde_kws={'linewidth':3, 'color':'k'},label="Median",)
    plt.legend()
```

```python
for i,var in enumerate(missing_num_var):
    plt.figure(figsize=(10,10))
    plt.subplot(3,1,1)
    sns.boxplot(df[var])
    plt.subplot(3,1,2)
    sns.boxplot(df4_num_mean[var])
    plt.subplot(3,1,2)
    sns.boxplot(df5_num_median[var])
    
```
