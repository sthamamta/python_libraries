```python
df7= pd.read_csv('data/Fortune_12.csv')
df7
```

```python
#using dropna()  : drops all rows with null value
df7.dropna()
```

```python
#using dropna(axis=1)  : drops all columns with null value
df7.dropna(axis=1)
```

```python
#drops the rows when any single null value is present in it
df7.dropna(how='any')
```

```python
# drops the rows with null value only whem all the value in a row are null values
df7.dropna(how='all')
```

```python
#using threshold parameter
df7
```

```python
#this displays row with atleast 7 not null values
df7.dropna(thresh = 7)
```

```python
# subset parameter
df7
```

```python
df7.dropna(subset=['Industry'])  
# it returns the dataframe which consists of rows where the value in Industry column are not null
```

```python
df7.dropna(subset=['Growth'])
```

```python
#using inplace parameter
df7.dropna(inplace=True)
#inplace parameter held true will make the changes in actual dataframe without creating new dataframe
```
