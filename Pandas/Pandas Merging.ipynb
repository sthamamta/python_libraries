```python
import pandas as pd
```

```python
# pandas merge connects columns or indexes in DataFrame based on one or more keys
```

```python
df1=pd.DataFrame({'ID':[1,2,3,4,5],'Class':[9,10,11,12,13]})
df1
```

```python
df2 = pd.DataFrame({'ID':[1,2,3,4,5],'Name':['A','B','C','D','E']})
df2
```

```python
pd.merge(df1,df2)
```

```python
pd.merge(df1,df2,on='ID')
```

```python
pd.merge(df2,df1,on='ID')
```

```python
df3=pd.DataFrame({'ID':[1,2,3,4,6],'Class':[9,10,11,12,13]})
df3
df4=pd.DataFrame({'ID':[1,2,3,4,5],'Class':[9,10,11,12,13]})
df4
```

```python
pd.merge(df3,df4)
```

```python
pd.merge(df3,df4,on='Class')
```

```python
pd.merge(df4,df3,on='Class')
```

```python
pd.merge(df3,df4,on='ID',how='right')
```

```python

df5 = pd.DataFrame({'ID':[1,2,3,4,6],'Name':['A','B','C','D','E']})
df6=pd.DataFrame({'ID':[1,2,3,4,5],'Class':[9,10,11,12,13]})
```

```python
pd.merge(df5,df6,on='ID',how='right')
```

```python
pd.merge(df5,df6,on='ID',how='left')
```

```python
pd.merge(df5,df6,on='ID',how='outer')
```

```python
pd.merge(df5,df6,on='ID',how='outer',indicator=True)
```

```python
df7 = pd.DataFrame({'ID':[1,2,3,4,6],'Name':['A','B','C','D','E']})
df8=pd.DataFrame({'ID':[7,8,9,10,11],'Class':[9,10,11,12,13]})
```

```python
pd.merge(df7,df8)
```

```python
pd.merge(df7,df8,left_index=True,right_index= True)
```

```python
df9 = pd.DataFrame({'ID':[1,2,3,4,6],'Name':['A','B','C','D','E']})
df10 = pd.DataFrame({'ID':[1,2,3,4,6],'Name':['A','B','C','D','E']})
pd.merge(df9,df10,on='ID')
```

```python
pd.merge(df9,df10,on='ID',suffixes=('_df9','_df10'))
```
