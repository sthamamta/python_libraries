# how to write csv file

```python
import pandas as pd
```

```python
df=pd.read_csv('Fortune_10.csv')
df
```

```python
type(df)
```

```python
#print the columns names
df.columns
```

```python
# reading only first row
df1=pd.read_csv('Fortune_10.csv',nrows=1)
df1
```

```python
# reading only first  5 rows
df2=pd.read_csv('Fortune_10.csv',nrows=5)
df2
```

```python
#printing columns with zero and four index
df3 = pd.read_csv('Fortune_10.csv',usecols=[0,4])
df3
```

```python
df4 = pd.read_csv('Fortune_10.csv',usecols=[2,4,7])
df4
```

```python
df5 = pd.read_csv('Fortune_10.csv',skiprows=1) #skips the first one row
df5
```

```python
df6 = pd.read_csv('Fortune_10.csv',skiprows=[2,3]) #skips one row with index 2 and 3
df6
```

```python
df7 = pd.read_csv('Fortune_10.csv',index_col = 'ID') #creates df with ID column as index
df7
```

```python
df8 = pd.read_csv('Fortune_10.csv',index_col = 3)
df8
```
