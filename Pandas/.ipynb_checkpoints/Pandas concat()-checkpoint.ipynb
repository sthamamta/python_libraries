```python
ser1=pd.Series([0,1,2,3])
ser1
```

```python
ser2=pd.Series([4,5,6])
ser2
```

```python
pd.concat([ser1,ser2])
```

```python
df = pd.DataFrame({'ID':[1,2,3,4],'Name':['A','B','C','D'],'Class':[11,12,13,14]})
df
```

```python
df1 = pd.DataFrame({'ID':[5,6,7,8],'Name':['e','f','g','h'],'Class':[17,13,14,16]})
df1
```

```python
pd.concat([df,df1])
```

```python
pd.concat([df1,df])
```

```python
pd.concat([df1,df],axis=1)
```

```python
pd.concat([df1,df],axis=1,ignore_index=True)
```

```python
pd.concat([df1,df],axis=0,ignore_index= True)
```

```python
pd.concat([df1,df],axis=0)
```

```python
df3 = pd.DataFrame({'ID':[1,2,3,4],'Name':['A','B','C','D'],'Class':[11,12,13,14]})
df3
df4 = pd.DataFrame({'ID':[3,4],'Name':['C','D'],'Class':[13,14]})
df4
```

```python
pd.concat([df3,df4],ignore_index=True)
```

```python
pd.concat([df3,df4],axis=1)
```

```python
#using join
pd.concat([df3,df4],axis=1,join='inner')
```

```python
pd.concat([df3,df4],axis=1,join='outer')
```

```python
pd.concat([df3,df4],axis=0,join='inner')
```

```python

pd.concat([df3,df4],axis=1)
```

```python
#using keys
df5= pd.DataFrame({'ID':[1,2,3,4],'Name':['A','B','C','D'],'Class':[11,12,13,14]})
df6= pd.DataFrame({'ID':[5,6,7,8],'Name':['e','f','g','h'],'Class':[15,16,17,18]})
pd.concat([df5,df6],keys=['df5','df6'])
```

```python
pd.concat([df5,df6],keys=['first df','second df'])
```

```python
pd.concat([df5,df6],keys=['first df','second df'],axis=1)
```

```python
df7= pd.DataFrame({'ID':[5,6,7,8],'Name':['e','f','g','h'],'Class':[15,16,17,18]})
df8= pd.DataFrame({'Marks':[11,44,55,66]})
```

```python
pd.concat([df7,df8])
```

```python
pd.concat([df7,df8],axis=1)
```
