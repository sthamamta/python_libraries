```python
import pandas as pd
```

```python
# fillna()
# fillna() fills the NaN values with given values(input)
```

```python
# value
# method
# axis
# limit
# inplace
# dictionary
```

```python
df =pd.read_csv('data/Fortune_12.csv')
df
```

```python
# filling 0 to NaN values
df.fillna(0)
```

```python
df.fillna(2)
```

```python
#using dictionary
df.fillna({'Industry':'none','Profit':'0','Growth':'0'})
```

```python
#fills NaN with its preceding values
df.fillna(method='ffill')
```

```python
#fills NaN with its suceeding values
df.fillna(method='bfill')
```

```python
#using axis
df.fillna(axis=0, method='ffill')
```

```python
#using axis
#fills NaN according to previous column value
df.fillna(axis=1, method='ffill')
```

```python
#limit
#limits the no of NaN values to be filled 
df.fillna(0,limit=2)
```

```python
#limit and method
#limits the no of NaN values to be filled 
#here two columns with NaN values are filled with its sucessor value
df.fillna(limit=2,method='ffill')
```

```python
#makes changes to original df without creating new df
df.fillna(0,inplace= True)
df
```
