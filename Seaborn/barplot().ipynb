```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
df = sns.load_dataset("tips")
df
```

```python
sns.barplot()
plt.plot()
```

```python
sns.barplot(x=df['total_bill'])
```

```python
sns.barplot(x=df['day'],y=df['total_bill'])
plt.show()
```

```python
#using hue
sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'])
plt.show()
```

```python
#using data parameter
sns.barplot(x='day',y='total_bill',hue='sex',data=df)
plt.show()
```

```python
#using order parameter
order=['Sun','Thur','Fri','Sat']
sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'],order=order)
plt.show()
```

```python
#using hue_order parameter
order=['Sun','Thur','Fri','Sat']
hue_order=['Female','Male']
sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'],order=order,hue_order=hue_order)
plt.show()
```

```python
#using estimator parameter
order=['Sun','Thur','Fri','Sat']
sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'],order=order,estimator=np.max)
plt.show()
```

```python
# using confidence interval(cf) parameter : effects the length of blackline at the top of bar
order=['Sun','Thur','Fri','Sat']
sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'],order=order,estimator=np.max,ci=50)
plt.show()
```

```python
#using orient parameter : cant be horizontal when x axis has categorical value
order=['Sun','Thur','Fri','Sat']
sns.barplot(x=df['total_bill'],y=df['day'],hue=df['sex'],order=order,estimator=np.max,ci=50)
plt.show()
```

```python
#using color parameter
order=['Sun','Thur','Fri','Sat']
sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'],order=order,color='m')
plt.show()
```

```python
#using palette parameter
order=['Sun','Thur','Fri','Sat']
sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'],order=order,palette='magma')
plt.show()
```

```python
#using saturation parameter
order=['Sun','Thur','Fri','Sat']
sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'],order=order,saturation=.4)
plt.show()
```

```python
#using errcolor parameter
order=['Sun','Thur','Fri','Sat']
sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'],order=order,palette='magma',errcolor="0.5")
plt.show()
```

```python
#using errwidth parameter
order=['Sun','Thur','Fri','Sat']
sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'],order=order,palette='magma',errcolor="0.5",errwidth=12)
plt.show()
```

```python
#using dodge parameter
order=['Sun','Thur','Fri','Sat']
sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'],order=order,dodge=False)
plt.show()
```

```python

order=['Sun','Thur','Fri','Sat']
sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'],order=order)
plt.show()
```

```python
#using kwargs
sns.set()
kwargs={'alpha':0.6,'linestyle':'--','linewidth':2,'edgecolor':'m'}
sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'],**kwargs)
plt.show()
```

```python
sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'],alpha=0.8,linestyle="--",linewidth=3,edgecolor='r')
plt.show()
```

```python
#using axis parameter

ax=sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'],alpha=0.8,linestyle="--",linewidth=3,edgecolor='r')
ax.set(title="the barplot of tips dataset",
      xlabel='days',
      ylabel='total bill')
plt.show()
```

```python
#using figsize
plt.figure(figsize=(16,9))
ax=sns.barplot(x=df['day'],y=df['total_bill'],hue=df['sex'],alpha=0.8,linestyle="--",linewidth=2,edgecolor='m')
plt.title('the barplot of tips dataset',fontsize=20)
plt.xlabel('days',fontsize=15)
plt.ylabel('total bill',fontsize=15)
plt.savefig('the barplot of tips dataset')
plt.show()
```
