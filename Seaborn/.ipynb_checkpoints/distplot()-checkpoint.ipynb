```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from scipy.stats import norm # for scientific Computing
```

```python
df = sns.load_dataset("tips")
df
```

```python
#plotting histogram
sns.distplot(df['size'])
plt.show()
```

```python
sns.distplot(df['total_bill'])
plt.show()
```

```python
sns.distplot(df['tip'])
plt.show()
```

```python
#using bins parameter
sns.distplot(df['tip'],bins=100)
plt.show()
```

```python
sns.distplot(df['tip'],bins=1)
plt.show()
```

```python
#using hist parameter : setting it false to hide histogram
sns.distplot(df['tip'],hist=False)
plt.show()
```

```python
#kde :kernal density estimate
sns.distplot(df['tip'],kde=False)
plt.show()
```

```python
#using  fit parameter
sns.distplot(df['tip'],fit=norm,kde=False)
plt.show()
```

```python
# using color parameter
sns.distplot(df['tip'],fit=norm,kde=False,color="r")
plt.show()
```

```python
# setting verticle parameter to true
sns.distplot(df['tip'],fit=norm,kde=False,vertical= True)
plt.show()
```

```python
#using axlabel parameter
sns.distplot(df['tip'],fit=norm,kde=False,axlabel='total tip')
plt.show()
```

```python
#using label and legend and title
sns.distplot(df['tip'],fit=norm,kde=False,axlabel='total tip',label="total tips in a restaurant")
plt.legend()
plt.title('total tips histogram',fontsize=18)
plt.show()
```

# Part 2

```python
#uisng sns.set()
plt.figure(figsize=(16,9))  # increase the size of plot
sns.set()
sns.distplot(df['total_bill'],fit=norm,kde=False,axlabel='total bill',label="total bill in a restaurant")
plt.legend()
plt.title('total bill histogram',fontsize=18)
plt.show()
```

```python
df.total_bill.sort_values()
```

```python
#using bins parameter

bins = [1,5,15,20,25,30,35,40,45,50,55]

plt.figure(figsize=(16,9))  # increase the size of plot
sns.set()

sns.distplot(df['total_bill'],fit=norm,kde=False,axlabel='total bill',label="total bill in a restaurant",bins=bins)
plt.legend()
plt.xticks(bins)

plt.title('total bill histogram',fontsize=18)
plt.show()
```

```python
# using hist_kws parameter

plt.figure(figsize=(16,9))  # increase the size of plot
sns.set() # for backgroud grid

sns.distplot(df['total_bill'],fit=norm,kde=False,axlabel='total bill',
             hist_kws={'color':'r','edgecolor':'k','linewidth':3,'linestyle':'--','alpha':0.9})


plt.title('total bill histogram',fontsize=18)
plt.show()
```

```python

plt.figure(figsize=(16,9))  # increase the size of plot
sns.set() # for backgroud grid

sns.distplot(df['total_bill'],fit=norm,axlabel='total bill',
             hist_kws={'color':'r','edgecolor':'k','linewidth':3,'linestyle':'--','alpha':0.9},
             kde_kws={'color':'b','linewidth':10,'linestyle':'--','alpha':0.9}
            )


plt.title('total bill histogram',fontsize=18)
plt.show()
```

```python
plt.figure(figsize=(16,9))  # increase the size of plot
sns.set() # for backgroud grid

sns.distplot(df['total_bill'],fit=norm,axlabel='total bill',
             hist_kws={'color':'y','edgecolor':'k','linewidth':3,'linestyle':'--','alpha':0.9},
             kde_kws={'color':'b','linewidth':10,'linestyle':'--','alpha':0.9},
             rug=True,
            rug_kws={'color':'m','linewidth':1,'linestyle':'--','alpha':0.9},
            )


plt.title('total bill histogram',fontsize=18)
plt.show()
```

```python
#using fit_kws parameter by setting kde to False
plt.figure(figsize=(16,9))  # increase the size of plot
sns.set() # for backgroud grid

# alpha paramter for transpaency
# fit=norm for showing normal distribution curve

sns.distplot(df['total_bill'],fit=norm,axlabel='total bill',kde=False,
             hist_kws={'color':'y','edgecolor':'k','linewidth':3,'linestyle':'--','alpha':0.9},
             fit_kws={'color':'b','linewidth':10,'linestyle':'--','alpha':0.9},
             rug=True,
            rug_kws={'color':'m','linewidth':1,'linestyle':'--','alpha':0.9},
            )


plt.title('total bill histogram',fontsize=18)
plt.show()
```

```python
# Plot multiple seaborn histogram in single graph
plt.figure(figsize=(16,9))
sns.distplot(df["total_bill"], bins=100, label="total_bil")
sns.distplot(df["tip"], bins=9, label="tip")
sns.distplot(df["size"], bins=9, label = "size")
 
plt.legend()
plt.show()
```
