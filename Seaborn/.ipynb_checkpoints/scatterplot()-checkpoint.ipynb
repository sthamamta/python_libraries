```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
titanic_df = sns.load_dataset('titanic')
titanic_df
```

```python
titanic_df.shape
```

```python
titanic_df.size
```

```python
titanic_df.isnull().size
```

```python
titanic_df.columns
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df)
plt.show()
```

```python
#using hue parameter
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='sex')
plt.show()
```

```python
#using hue parameter
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='sex',style='who')
plt.show()
```

```python
#using size parameter
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='sex',style='who',size='who')
plt.show()
```

```python
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='sex',style='who',size='who',sizes=(100,400))
plt.show()
```

```python
#increasing the fig size
plt.figure(figsize=(16,9))
sns.scatterplot(x='age',y='fare',data=titanic_df,hue='sex',style='who',size='who',sizes=(100,400))
plt.show()
```

```python
#increasing the fig size
plt.figure(figsize=(16,9))
sns.scatterplot(x='who',y='fare',data=titanic_df,hue='alive',style='alive',size='who',sizes=(100,400))
plt.show
#here women who paid higher fare were mostly alive
```

```python
#iusing palette parameter
plt.figure(figsize=(16,9))
sns.scatterplot(x='who',y='fare',data=titanic_df,hue='alive',style='alive',size='who',sizes=(100,400),
               palette='inferno_r',alpha=0.7)
plt.show
```
