```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# heatmap is a graphical representation of 2d data
#each value represents in a matrix and it has a special color
```

```python
arr_2d = np.linspace(1,5,12).reshape(4,3)
arr_2d
```

```python
sns.heatmap(arr_2d)
plt.plot()
```

```python
# Load dataset
 
globalWarming_df = pd.read_csv("Who_is_responsible_for_global_warming.csv")
globalWarming_df.head()
```

```python
# set country name as index and drop Country Code, Indicator Name and Indicator Code
 
globalWarming_df  = globalWarming_df.drop(columns=['Country Code', 'Indicator Name', 'Indicator Code'], axis=1).set_index('Country Name')
globalWarming_df
```

```python
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df)
plt.show
```

```python
#using vmin and vmax parameter
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,vmin=0,vmax=50) #sets maximum and minimum value on the color scale
plt.show
```

```python
#changing the color of heatmap using cmap parameter
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,cmap='coolwarm')
plt.show()
```

```python
#displaying color value by setting annot parameter to True
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,annot=True)
plt.show()
```

```python
arr_2d
```

```python
# fmp parameter - add text on heatmap cell
 #The seaborn heatmap fmt helps to show annot with different formatting.
annot_arr = np.array([['a00','a01','a02'],
                      ['a10','a11','a12'],
                      ['a20','a21','a22'],
                      ['a30','a31','a32']], dtype = str)
 
sns.heatmap(arr_2d, annot = annot_arr, fmt="s") # s -string, d - decimal
plt.show()
```

# Part 2

```python
#using annot_kws parameter
plt.figure(figsize=(16,9))    
annot_kws={'fontsize':10,
          'fontstyle':'italic',
          'color':'k',
          'backgroundcolor':'w',
          'rotation':'vertical',
          'verticalalignment':'center'}
sns.heatmap(globalWarming_df,cmap='coolwarm',annot=True,annot_kws=annot_kws)
plt.show()
```

```python
#using linewidth parameter
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,cmap='coolwarm',linewidth=10,annot=True)
plt.show()
```

```python
#changing the line color

plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,cmap='coolwarm',linewidth=10,annot=True,linecolor='m')
plt.show()
```

```python
#hiding the colorbar by setting cbar parameter to False
#using linewidth parameter
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,cmap='coolwarm',linewidth=5,annot=True,cbar=False)
plt.show()
```

```python
#hiding xlabels and ylabels
#using linewidth parameter
plt.figure(figsize=(16,9))
sns.heatmap(globalWarming_df,cmap='coolwarm',linewidth=10,annot=True,xticklabels=False,yticklabels=False)
plt.show()
```

```python
#using cbar_kws parameter to change the colorbar into various forms
plt.figure(figsize=(16,9))
cbar_kws = {"orientation":"horizontal", 
            "shrink":1,
            'extend':'min', 
            'extendfrac':0.1, 
            "ticks":np.arange(0,22), 
            "drawedges":True,
           }
 
sns.heatmap(globalWarming_df,cmap='coolwarm',cbar_kws=cbar_kws)
plt.show()
```

```python
# change y-axis labels using yticklabels parameter
 
plt.figure(figsize=(16,9))
 
country_code = ['USA', 'GBR', 'IND', 'CHN', 'RUS', 'AUS', 'FRA', 'DEU', 'CAN', 'BRA', 'ARG', 'PAK', 'NPL', 'BGD', 'JPN']
 
sns.heatmap(globalWarming_df, yticklabels = country_code,xticklabels=np.arange(0,15))
plt.show()
```

```python
plt.figure(figsize=(16,9))

ax= sns.heatmap(globalWarming_df)
ax.set(
    title='Heatmap of who is responsible for globalwarming',
    xlabel='CO2 emission for year',
    ylabel='country name'
)
sns.set(font_scale=2)  #increase the size of label
plt.show()
```
